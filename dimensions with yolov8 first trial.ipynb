```python
# Install YOLOv8 dependencies
!pip install ultralytics --quiet

from ultralytics import YOLO
import os
import shutil
import cv2
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import math

# Set dataset paths
DATASET_PATH = '/kaggle/input/heart-gan-bg-removed/processed_heart_images'
WORK_DIR = '/kaggle/working/heart_dataset'
IMG_DIR = f'{WORK_DIR}/images'
LABEL_DIR = f'{WORK_DIR}/labels'

# Create YOLOv8 folder structure
for folder in ['images/train', 'images/val', 'labels/train', 'labels/val']:
    os.makedirs(os.path.join(WORK_DIR, folder), exist_ok=True)

# Dummy label generator (center of image as landmark)
image_list = sorted(os.listdir(DATASET_PATH))
train_files, val_files = train_test_split(image_list, test_size=0.2, random_state=42)

def create_dummy_label(img_name, split):
    img_path = os.path.join(DATASET_PATH, img_name)
    img = cv2.imread(img_path)
    h, w = img.shape[:2]

    # Compute center point of image (this simulates a landmark)
    cx, cy = w // 2, h // 2

    # YOLO format: class x_center y_center width height (normalized)
    x_center = cx / w
    y_center = cy / h
    box_w = box_h = 0.05  # Small bbox around landmark
    label = f"0 {x_center:.6f} {y_center:.6f} {box_w:.6f} {box_h:.6f}\n"

    split_label_dir = os.path.join(WORK_DIR, f'labels/{split}')
    with open(os.path.join(split_label_dir, img_name.replace('.png', '.txt')), 'w') as f:
        f.write(label)

    # Copy image
    shutil.copy(img_path, os.path.join(WORK_DIR, f'images/{split}', img_name))

# Generate dummy labels
for img in train_files:
    create_dummy_label(img, 'train')
for img in val_files:
    create_dummy_label(img, 'val')

# Create data.yaml file for YOLOv8
with open(f'{WORK_DIR}/data.yaml', 'w') as f:
    f.write(
        f"path: {WORK_DIR}\n"
        f"train: images/train\n"
        f"val: images/val\n"
        f"names:\n  0: landmark"
    )

# Train YOLOv8
model = YOLO("yolov8n.pt")  # use small model for speed, or yolov8m/l for better accuracy
results = model.train(
    data=f"{WORK_DIR}/data.yaml",
    epochs=10,  # You can increase this
    imgsz=256,
    batch=16,
    project='/kaggle/working/',
    name='yolo_heart_landmarks',
    verbose=True
)

# Load trained model
trained_model = YOLO("/kaggle/working/yolo_heart_landmarks/weights/best.pt")

# Run prediction on validation images
preds = trained_model.predict(source=os.path.join(WORK_DIR, 'images/val'), save=False, conf=0.5)

# Extract x, y, z from predictions
x_val, y_val, z_val, image_names = [], [], [], []

for i, result in enumerate(preds):
    img_name = os.path.basename(result.path)
    boxes = result.boxes
    if boxes is not None and len(boxes.xywh) > 0:
        x_center, y_center, _, _ = boxes.xywh[0].cpu().numpy()
        x, y = int(x_center), int(y_center)
        z = int(math.sqrt(x**2 + y**2))

        x_val.append(x)
        y_val.append(y)
        z_val.append(z)
        image_names.append(img_name)

# Create CSV
df = pd.DataFrame({
    'image_name': image_names,
    'x': x_val,
    'y': y_val,
    'z': z_val
})
csv_path = "/kaggle/working/heart_dimensions.csv"
df.to_csv(csv_path, index=False)

# Display sample predictions
import random
sample_ids = random.sample(range(len(image_names)), 5)

for idx in sample_ids:
    img_path = os.path.join(WORK_DIR, 'images/val', image_names[idx])
    img = cv2.imread(img_path)
    x, y, z = x_val[idx], y_val[idx], z_val[idx]

    cv2.circle(img, (x, y), 6, (0, 255, 0), -1)
    cv2.putText(img, f"X:{x}, Y:{y}, Z:{z}", (x+10, y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 1)

    plt.figure(figsize=(4, 4))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(image_names[idx])
    plt.axis('off')
    plt.show()

# Plot accuracy and loss curves
metrics_path = "/kaggle/working/yolo_heart_landmarks/results.csv"
if os.path.exists(metrics_path):
    results_df = pd.read_csv(metrics_path)
    plt.plot(results_df['      metrics/precision(B)'], label='Precision')
    plt.plot(results_df['      metrics/recall(B)'], label='Recall')
    plt.plot(results_df['      metrics/mAP50(B)'], label='mAP@0.5')
    plt.title("Validation Metrics per Epoch")
    plt.legend()
    plt.show()
```
